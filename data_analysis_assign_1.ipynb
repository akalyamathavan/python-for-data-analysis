```python
import numpy as np

## create 1D Numpy Array:

temperatures_w1 = np.array([22.5,25.3,20.8,23.4,26.1,24.8,21.9])
print("Week 1 Temperatures:",temperatures_w1)
```

```python
## Inspection and properties:

print("Shape:", temperatures_w1.shape)
print("Data Type:", temperatures_w1.dtype)
print("Number of Elements:", temperatures_w1.size)
```

```python
##  Array operations:

fahrenheit = (temperatures_w1 * 9/5) + 32
print("Temperatures in Fahrenheit:", fahrenheit)

print("Maximum:", np.max(temperatures_w1))
print("Minimum:", np.min(temperatures_w1))
print("Mean:", np.mean(temperatures_w1))
```

```python
##  Array slicing and indexing:

print("First 3 days:", temperatures_w1[:3])
print("Weekend (last 2 days):", temperatures_w1[-2:])
print("Middle 3 days:", temperatures_w1[2:5])
```

```python
##  Create 2D array:

temperatures = np.array([
    [22.5, 25.3, 20.8, 23.4, 26.1, 24.8, 21.9],   # Week 1
    [19.2, 22.5, 21.3, 24.0, 23.5, 22.8, 20.1]    # Week 2
])

print("2D Temperature Array:\n", temperatures)

##  Inspect and slice the 2D array:

print("\nShape:", temperatures.shape)
print("Data Type:", temperatures.dtype)
print("Total Elements:", temperatures.size)

# Weekend temperatures for both weeks
print("Weekend Temps (Both Weeks):")
print(temperatures[:, -2:])
```

```python
## PANDAS SERIES :
import pandas as pd

## Create pandas series

marks = pd.Series(
    [95, 92, 89, 85, 80],
    index=['Rank1', 'Rank2', 'Rank3', 'Rank4', 'Rank5']
)
print(marks)
```

```python
##  Indexing and slicing

print("1st Rank Mark:", marks.iloc[0])
print("Top 3 Ranks:\n", marks.loc[['Rank1', 'Rank2', 'Rank3']])
print("3rd Rank Mark:", marks.iloc[2])

print("Marks > 90:\n", marks[marks > 90])
```

```python
## Manipulating Series:

marks = pd.Series(
    [95, 92, 89, 85, 80],
    index=['Rank1', 'Rank2', 'Rank3', 'Rank4', 'Rank5']
)

marks['Rank1'] = 100
marks = marks.drop('Rank5')

cgpa = marks / 10
print("Updated Marks:\n", marks)
print("CGPA:\n", cgpa)
```

```python
## PANDAS DATAFRAME :

##   Creating Dataframe:
transactions = pd.DataFrame({
    'TransactionID': [101,102,103,104,105,106,107,108,109,110],
    'ProductCategory': ['Electronics','Clothing','Electronics','Furniture',
                        'Clothing','Electronics','Furniture','Clothing',
                        'Furniture','Electronics'],
    'Region': ['North','South','North','East','West',
               'North','East','West','South','North'],
    'Amount': [200,150,300,450,200,250,300,180,350,400]
})

print(transactions)
```

```python
## Data Exploration :

print(transactions.head())
print(transactions.tail())
print("Shape:", transactions.shape)
print("Columns:", transactions.columns)
print(transactions.dtypes)

print(transactions[['ProductCategory', 'Amount']])
print(transactions.iloc[:, -3:])

print(transactions[(transactions['Region'] == 'North') & (transactions['Amount'] > 200)])

print(transactions['ProductCategory'].value_counts())
print(transactions['Region'].unique())

print(transactions.groupby('Region')['Amount'].mean())
```

```python
##Manipulating the Dataframe :

transactions.loc[transactions['TransactionID'] == 102, 'Amount'] = 165

transactions['Discount'] = transactions['Amount'] * 0.10

transactions = transactions[transactions['TransactionID'] != 109]

transactions.drop(columns=['Discount'], inplace=True)

print("Final DataFrame:\n", transactions)
```
